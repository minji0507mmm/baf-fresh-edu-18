#### 5. 분류 모델의 성능을 평가할 때 Precision과 Recall의 차이점은?

Precision이 중요한 상황과 Recall이 중요한 상황을 각각 적어주세요.

- precision은 모델이 "양성이라고 예측한 것"중에서 실제로 양성인 비율을 뜻하며 스팸 메일 필터, 무고한 사람을 범죄자로 분류하는 시스템 등에서 중요하다.(잘못 양성으로 판단하면 문제가 되는 상황)
- recall은 실제 양성 중에서 모델이 맞게 찾아낸 비율을 뜻하며 암 진단, 당뇨병 예측, 사기 탐지 시스템에서 중요하다.(실제 양성을 놓치면 문제가 되는 상황)


#### 6. 결측치 처리 방법에는 여러 가지가 있습니다. 아래와 같은 상황에서는 어떤 결측치 처리 방법이 가장 적절한가요?

- 데이터에서 일부 숫자형 변수의 결측치 비율이 40% 이상이다.
- 데이터를 분석해보니 결측값이 특정한 패턴 없이 무작위로 발생했다.

- 결측치 비율이 40% 이상으로 매우 높기 때문에 해당 변수를 단순히 제거하는 것은 데이터 손실이 크고 중요한 정보까지 잃을 위험이 있다. 또한 평균이나 중앙값과 같은 단순 통계량으로 대체할 경우 데이터 분포가 왜곡되고 분산이 감소하여 모델의 성능 저하를 초래할 수 있다.
- 다른 변수들과의 관계를 활용하여 결측값을 예측하는 방법이 통계적으로 타당하다. 따라서 단순 제거 또는 평균,중앙값 대체보다는 회귀분석 기반의 모델을 학습하여 결측값을 예측하는 모델 기반 결측치 대체 방법을 사용하는 것이 가장 적절하다.

#### 7. K-means 알고리즘의 단점 중 아래와 같은 상황에서 어떤 방법을 적용할 수 있을까요?

거리 기반 알고리즘으로 속성(피쳐)의 개수가 많은 경우 정확도가 떨어졌다.

- K-means는 거리 기반 알고리즘이므로 피쳐 수가 많아질 경우 차원의 저주로 인해 거리 계산의 구분력이 감소하여 정확도가 떨어질 수 있다. 이를 해결하기 위해 PCA와 같은 차원 축소 기법을 적용하여 주요 정보를 유지하면서 차원을 줄인 후 K-means를 수행하는 것이 적절하다.

#### 8. 앙상블 학습(Ensemble Learning) 방법 중 Bagging과 Boosting의 차이점

각각 대표적인 알고리즘을 하나씩 제시해주세요.

- Bagging은 부트스트랩 샘플링을 통해 여러 모델을 병렬로 학습한 후 평균 또는 투표 방식으로 결합하여 분산을 줄이는 방법이다. 대표적으로 Random Forest가 있다.
- Boosting은 이전 모델의 오차를 보완하도록 순차적으로 모델을 학습하며 편향을 줄이는 데 목적이 있다. AdaBoost와 Gradient Boosting이 대표적인 예이다.

#### 9. 다중공선성이란 무엇이며, 다중공선성이 심할 경우 해결 방법은 무엇인가요?



- 다중공선성이란 독립변수들 간에 강한 상관관계가 존재하는 현상을 의미한다. 이는 회귀계수의 불안정성을 증가시키고 해석을 어렵게 만든다. 다중공선성이 심할 경우 상관성이 높은 변수 중 하나를 제거하거나 PCA와 같은 차원 축소 기법을 사용하거나 Ridge 또는 Lasso와 같은 규제 기법을 적용하여 해결할 수 있다.

#### 10. 머신러닝 모델의 과적합(Overfitting)을 방지하기 위한 방법을 2가지 이상 적어주세요.

- 과적합을 방지하기 위해 규제 기법(Lasso, Ridge)을 사용하여 모델 복잡도를 줄일 수 있으며 교차검증을 통해 일반화 성능을 평가할 수 있다. 또한 트리의 깊이를 제한하거나 파라미터 수를 감소시키는 방법도 과적합을 줄이는 데 효과적이다.

```python
log_model = Pipeline([
    ('scaler', StandardScaler()),
    ('logistic', LogisticRegression(max_iter=1000, random_state=42))
])

log_model.fit(X_tr, y_tr)

log_pred = log_model.predict(X_val)
log_proba = log_model.predict_proba(X_val)[:, 1]

print("Logistic Regression 성능")
print("Accuracy:", accuracy_score(y_val, log_pred))
print("ROC-AUC:", roc_auc_score(y_val, log_proba))
print(classification_report(y_val, log_pred))
```
